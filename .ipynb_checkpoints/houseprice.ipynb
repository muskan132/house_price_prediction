```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
for feature in cafeatures:
    tmp = hd.copy()
    tmp.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

```python
for feature in disfeatures:
    tmp = hd.copy()
    tmp.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

```python
for feature in confeatures:
    data=hd.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
#since we caan see a lot of varriance and outliers we have to take the log of data and handle outliers
```

```python
import numpy as np
```

```python
## We will be using logarithmic transformation


for feature in confeatures:
    data=hd.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

```python
#outliers
for feature in confeatures:
    data=hd.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
